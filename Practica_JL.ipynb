```
import numpy as np
#os hago el trabajo más duro
from google.colab import files
import io

# cargar
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

# pandas
import pandas as pd

df = pd.read_csv(io.StringIO(uploaded['train_sentiment_utf8.csv'].decode('utf-8')))
```

```
documents = df['SentimentText']
```

```
# Generamos el vocabulario
def generate_vocabulary_maps(docs):
    vocabulary = {}
    inverse_vocabulary = {}
    for doc in documents:
        for token in doc.split(' '):
            if token not in vocabulary:
                vocabulary[token] = len(vocabulary)
                inverse_vocabulary[len(inverse_vocabulary)] = token
    return vocabulary, inverse_vocabulary

# Cuenta el número de apariciones de un elelento por entrada
def itransform(X, vocab):
    assert type(X)==list, 'X must be a list'
    rows= len(X)
    #print(rows)
    rep = np.zeros((rows, len(vocab)), dtype='int32')
    for i, x in enumerate(X):
        tokens = [vocab[token] for token in x.split(' ') if token in vocab]
        for t in tokens:
            rep[i, t] += 1
    return rep
```

```
# 
vocabulary, inverse_vocabulary = generate_vocabulary_maps(documents)

#
new_representation_np = np.zeros((len(vocabulary)), dtype='int32')

# 
itransform(documents.values.tolist(), vocabulary)

```
